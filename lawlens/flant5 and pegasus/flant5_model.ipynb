```python
from datasets import load_dataset

try:
    dataset = load_dataset("d0r1h/ILC")
    print("Dataset loaded successfully!")
    print(f"\nColumns: {dataset['train'].column_names}")
    print("\nSample Case:", dataset["train"][0]["Case"][:200] + "...")
    print("\nSample Summary:", dataset["train"][0]["Summary"])
except Exception as e:
    print(f"Error loading dataset: {e}")
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

model_name = "google/flan-t5-base"
tokenizer = T5Tokenizer.from_pretrained(model_name, legacy=False)
model = T5ForConditionalGeneration.from_pretrained(model_name).to("cuda")

print(f"\nModel loaded ({model_name})")
print(f"Tokenizer vocab size: {tokenizer.vocab_size}")
```

```python
def preprocess_function(examples):
    inputs = ["summarize legal case: " + doc for doc in examples["Case"]]
    model_inputs = tokenizer(inputs, max_length=1024, truncation=True, padding="max_length")

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples["Summary"],
            max_length=256,
            truncation=True,
            padding="max_length"
        ).input_ids

    model_inputs["labels"] = labels
    return model_inputs

tokenized_dataset = dataset.map(preprocess_function, batched=True)
print("\nPreprocessed keys:", list(tokenized_dataset["train"][0].keys()))
```

```python
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
import numpy as np
from rouge_score import rouge_scorer

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)

    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
    scores = [scorer.score(pred, label) for pred, label in zip(decoded_preds, decoded_labels)]

    return {
        "rouge1": np.mean([s['rouge1'].fmeasure for s in scores]),
        "rougeL": np.mean([s['rougeL'].fmeasure for s in scores]),
    }

# Corrected training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./flan-t5-ilc",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    eval_strategy="epoch",  # Changed from evaluation_strategy
    save_strategy="epoch",
    fp16=True,
    learning_rate=3e-5,
    logging_steps=100,
    predict_with_generate=True,
    load_best_model_at_end=True  # Recommended for early stopping
)
```

```python
from transformers import Seq2SeqTrainer, DataCollatorForSeq2Seq

# 1. Verify tokenizer range first
print(f"Tokenizer valid ID range: 0-{tokenizer.vocab_size - 1}")

# 2. Data collator with extra safety checks
data_collator = DataCollatorForSeq2Seq(
    tokenizer,
    model=model,
    padding=True,
    max_length=1024,  # Match your preprocessing
    pad_to_multiple_of=8  # Helps with GPU efficiency
)

# 3. Clean dataset by filtering invalid tokens
def is_valid_example(example):
    return all(0 <= token_id < tokenizer.vocab_size
              for token_id in example["input_ids"] + example["labels"])

tokenized_dataset = tokenized_dataset.filter(is_valid_example)

# 4. Training setup
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    logging_dir="./logs",
    logging_steps=100,
    save_total_limit=2,
    #evaluation_strategy="epoch",
    save_strategy="epoch",
    predict_with_generate=True,
    report_to=[],  # Disables WandB
    fp16=True,
    gradient_accumulation_steps=2  # Helps with memory
)

# 5. Create trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset.get("test"),
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# 6. Start training
print("\nStarting training...")
try:
    trainer.train()
except Exception as e:
    print(f"Training failed: {str(e)}")
    print("Check your tokenized data for invalid token IDs")
```

```python
SystemSummary_train = []
from transformers import T5Tokenizer, T5ForConditionalGeneration

device = "cuda" if torch.cuda.is_available() else "cpu"

# Load FLAN-T5
model_name = "google/flan-t5-base"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name).to(device)

SystemSummary_train = []

# Generate summaries for first 100 train cases (excluding index 0)
for i, case in enumerate(trainCasesText[1:101]):
    input_text = "summarize: " + case
    batch = tokenizer(input_text, truncation=True, padding="longest", return_tensors="pt").to(device)
    summary = model.generate(**batch)
    summary_final = tokenizer.batch_decode(summary, skip_special_tokens=True)[0]

    SystemSummary_train.append(summary_final)

    print(f"Generated summary for train case {i+1}/100")
```

```python
from rouge import Rouge
import pandas as pd

# Convert gold summaries to list of strings (ensure correct slice)
trainGoldSummary_list = trainGoldSummary[1:101].tolist()

# Initialize ROUGE evaluator
rouge = Rouge()

# Filter out empty summaries and corresponding gold summaries
filtered_SystemSummary_train = []
filtered_trainGoldSummary_list = []
for hyp, ref in zip(SystemSummary_train, trainGoldSummary_list):
    if hyp and ref:  # Check if both are non-empty
        filtered_SystemSummary_train.append(hyp)
        filtered_trainGoldSummary_list.append(ref)

# Evaluate the generated summaries using the ROUGE metric (averaged over all non-empty)
if filtered_SystemSummary_train:  # Proceed only if non-empty summaries exist
    score_train_multiple = rouge.get_scores(
        filtered_SystemSummary_train, filtered_trainGoldSummary_list, avg=True
    )

    # Display results
    print("\nAverage ROUGE scores for the train data:")
    print(pd.DataFrame(score_train_multiple).iloc[0])
else:
    print("\nNo non-empty summaries were generated for evaluation.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Extract average ROUGE scores for training data
rouge_1_train_avg = score_train_multiple['rouge-1']['f']
rouge_2_train_avg = score_train_multiple['rouge-2']['f']
rouge_l_train_avg = score_train_multiple['rouge-l']['f']

# Evaluate the generated summaries for the test data using the ROUGE metric
score_test_multiple = rouge.get_scores(SystemSummary_test, testGoldSummary_list, avg=True)

# Extract average ROUGE scores for test data
rouge_1_test_avg = score_test_multiple['rouge-1']['f']
rouge_2_test_avg = score_test_multiple['rouge-2']['f']
rouge_l_test_avg = score_test_multiple['rouge-l']['f']

# Plotting average ROUGE scores
labels = ['ROUGE-1', 'ROUGE-2', 'ROUGE-L']
train_avg_scores = [rouge_1_train_avg, rouge_2_train_avg, rouge_l_train_avg]
test_avg_scores = [rouge_1_test_avg, rouge_2_test_avg, rouge_l_test_avg]

x = range(len(labels))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(x, train_avg_scores, color='blue')
plt.xlabel('ROUGE Metrics')
plt.ylabel('Average ROUGE Scores')
plt.title('Average ROUGE Scores for Train Data')
plt.xticks(x, labels)

plt.subplot(1, 2, 2)
plt.bar(x, test_avg_scores, color='red')
plt.xlabel('ROUGE Metrics')
plt.ylabel('Average ROUGE Scores')
plt.title('Average ROUGE Scores for Test Data')
plt.xticks(x, labels)

plt.tight_layout()
plt.show()
```
